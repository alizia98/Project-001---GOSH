```python
# import addmission data
ac_df = pd.read_csv("AdmissionsCorePopulatedTable.txt", sep='\t', infer_datetime_format = True, parse_dates = ['AdmissionStartDate', 'AdmissionEndDate'])
# ac_df.head()
```

```python
ac_df["AdmissionYear"] = pd.to_datetime(ac_df["AdmissionStartDate"], format='%Y-%m-%d %H:%M:%S.%f').dt.year

# AdmissionEndDate has some bad values here so won't convert to a datetime natively
ac_df["AdmissionEndDate"] = pd.to_datetime(ac_df["AdmissionEndDate"], format='%Y-%m-%d %H:%M:%S.%f', errors = 'coerce')

ac_df["DaysStayed"] =  ac_df["AdmissionEndDate"]- ac_df["AdmissionStartDate"]

ac_df["DaysStayed"] = ac_df["DaysStayed"].dt.days

ac_df = ac_df.drop(['AdmissionStartDate','AdmissionEndDate'], axis = 1)
```

```python
# drop nah and merge
pc_df['PatientID'] = pc_df['PatientID'].dropna()
ac_df['PatientID'] = ac_df['PatientID'].dropna()
acpc_df = pd.merge(ac_df, pc_df, how = 'left', left_on = ['PatientID'], right_on = ['PatientID'])
```

```python
# read diagnoses data
dc_df = pd.read_csv("AdmissionsDiagnosesCorePopulatedTable.txt", sep='\t')
dc_df = dc_df.drop(['PrimaryDiagnosisDescription'], axis = 1)
dc_df['PrimaryDiagnosisCode'] = dc_df.PrimaryDiagnosisCode.astype('category')
```
